##Finding out missing value percentage

```python
marketing_train.isnull()
```

```python
missing_val = pd.DataFrame(marketing_train.isnull().sum())
missing_val
```

```python
missing_val = missing_val.reset_index()
missing_val
```

```python
missing_val = missing_val.rename(columns = {'index' : 'Variables', 0 : 'Missing_percentage'})
missing_val
```

```python
missing_val['Missing_percentage'] = (missing_val['Missing_percentage']/len(marketing_train))*100
missing_val
```

```python
missing_val.to_csv("Missing_perc.csv", index = False)
```

```python
mv = missing_val.sort_values('Missing_percentage', ascending = False)
mv
```

```python
mv.to_csv("Sorted_missing_%.csv", index = False)
```

##Imputation

1. get value 
2. make it nan
3. impute (mean / median)
4. check value
  - select the value which is closest to the original value

```python
import numpy as np
```

```python
marketing_train['custAge'].loc[70]
```

```python
marketing_train['custAge'].loc[70] = np.nan
```

```python
marketing_train['custAge'].loc[70]
```

```python
marketing_train['custAge'] = marketing_train['custAge'].fillna(marketing_train['custAge'].mean())
```

```python
marketing_train['custAge'].loc[70]
```

> imputed value with median was less than mean so we choose median to impute. **bold text**
